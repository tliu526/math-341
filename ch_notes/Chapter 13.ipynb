#Chapter 13: Continous Random Variables -  Uniform and Exponential

##13.1 The Uniform Distribution

####Definition
We say that a random variable $X$ as the uniform distribution on $[a,b]$ with $-\infty < a < b < \infty)$ if:

$$f_X(x) = \begin{cases}
\frac{1}{b-a} & \text{if } a \le x \le b \\
0 & \text{otherwise.}
\end{cases}$$

We denote this by writing $X \sim \text{Unif}(a,b)$.

###13.1.1 Mean and Variance
Let $X \sim \text{Unif}(a,b)$. Then the mean $\mu_X$ and variance $\sigma_X^2$ are:

$$\mu_X = \frac{b+a}{2}$$

$$\sigma_X^2 = \frac{(b-a)^2}{12}$$

In the special case that $[a,b] = [0,1]$ the mean is 1/2 and the variance is 1/12, while if $[a,b] = [-1/2, 1/2]$ the mean is 0 and the variance is 1/12. Another important case is when $[a,b] = [-\sqrt{3}, \sqrt{3}]$, as in this case the mean is 0 and the variance is 1.

###13.1.2 Sums of Uniform Random Variables

Let $X$ and $Y$ be independent, Unif$(0,1)$ random variables. Then the density of $Z = X + Y$ is:

$$
f_Z(z) = \begin{cases}
z & \text{if } 0 \le z \le 1 \\
2-z & \text{if } 1 \le z \le 2 \\
0 & \text{otherwise.}
\end{cases}$$

The theory of convolutions (Chapter 10) gives the answer. __Symmetric arguments__ are very powerful and frequently can save lots of tedious, duplicate calculations.

##13.2 The Exponential Distribution

###Definition

We say a random variable $X$ has the exponential distribution with parameter $\lambda > 0$ if:

$$
f_X(x) = \begin{cases}
\frac{1}{\lambda}e^{-x/\lambda} & \text{if } x \ge 0 \\
0 & \text{otherwise.}
\end{cases}
$$

We denote this by $X \sim \text{Exp}(\lambda)$.

###13.2.1 Mean and Variance

Let $X \sim \text{Exp}(\lambda)$. Then the mean $\mu_X$ and variance $\sigma_X^2$ are:

$$\mu_X = \lambda$$
$$\sigma_X^2 = \lambda^2$$

In the special case $\lambda = 1$, the mean and the variance are both 1.

###13.2.2 Sums of Exponential Random Variables

###Definition (Erlang Distribution)
Let $X_1,...,X_n$ be $n$ independent identically distributed exponential random variables with parameter $\lambda > 0$. Then the density of $X = X_1 + ... + X_n$ is:

$$
f_X(x) = \begin{cases}
\frac{x^{n-1}e^{-x/\lambda}}{\lambda^n (n-1)!} & \text{if } x \ge 0 \\
0 & \text{otherwise.}
\end{cases}
$$

We say any random variable with this density $f_X$ is an Erlang distribution with parameters $\lambda$ and $n$; it has a mean $n\lambda$ and a variance of $n\lambda^2$.

###13.2.4 Generating Random Numbers from Exponential Distributions

###Definition (Cumulative Distribution Method to Generate Random Numbers)
Let $X$ be a random variable with density $f_X$ and cumulative distribution function $F_X$. If $Y$ is a uniform random variable $[0,1]$, then:

$$X = F_X^{-1}(Y)$$

This is also known as the __inverse transform sampling__ or the __inverse transformation method__.
