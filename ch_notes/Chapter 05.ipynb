###5.1.2 Choosing Groups

_Problem 5.1.1:_ Assume there are 12 people running for student government.

- How many ways are there to choose 4 of them? 

    $\binom{12}{4} = 495$

- How many ways are there to choose one of them to be presdient, one to be viec-president, one to be treasure and one to be secretary, given that no one can hold two positions? 

    $12 \cdot 11 \cdot 10 \cdot 9$ as order matters.

- How many ways are there to choose one of them to be president, one to be vice-president, one to be treasurer and one to be secretary, allowing people to hold any number of offices. 

    $12^4$ as each person can hold any number of positions.

###5.1.3 Circular Orderings

- How many ways are there for 5 people to line up, where order matters?

    $5! = 120$
    
- How many ways are there for 5 people to sit at a circular table, where all that matters is relative ordering?

    There is no first or last place in line. We should select one person to be special and rotate until they're at the (1,0) position. The there are $4!$ ways to order the remainig people after our special person.
    
In general, there are $(n-1)!$ different ways that $n$ people can arrange themselves around a circular table when order matters. 

###5.1.4 Choosing Ensembles 

_Problem 5.1.3:_ You have 5 different pairs of socks, 3 different pairs of pants, and 4 different apirs of shirts to choose from. Assume exactly two pairs of socks are red, none of the pants are red, and only one shirt is red. You must choose exactly one pair of socks, one pair of pants, and a shirt.

- How many different ensembles can you make?

    We want to figure out how many unique ways we can choose a pair of socks, pants, and a shirt to make an outfit. We just multiply out: $5 \cdot 3 \cdot 4 = 60$.
    
    
- How many different ensembles can you make where a least one item is red?

    We want either a pair of red socks or a red shirt (or both). First count how many outfits contain just a red pair of socks. Two red pairs of socks: $2 \cdot 3 \cdot 4 = 24$. Since there is only one red shirt: $5 \cdot 3 \cdot 1 = 15$. Note here that we have double counted outfits with both red socks and red shirts. So we calculate the number of outfits with both and subtract that out. Two pairs of red socks, one red shirt: $2 \cdot 3 \cdot 1 = 6$ outfits with both a red shirt and red socks. Thus the final amount is $24 - 15 - 6 = 33$. 
    
This method of counting is the _Inclusion-Exclusion Principle_.

####Important

Another approach: every outfit will either conatin a red element or not. We know 60 possible outfits, so we just need to find the outfits with _no_ red and subtract that amount. $3 \cdot 3 \cdot 3 = 27$. $60 - 27 = 33$ as generated by our first method.


####Definition (Complementary Events)

For many problems the easiest way to find the probability of $A$ is to find the probability that $A$ does not happen as $Pr(A) = 1 - Pr(A^c)$. Useful for problems with the phrase _at least one_.

##5.2 The Method of Inclusion-Exclusion

###5.2.1 Special Cases of the Inclusion-Exclusion Principle

- We use unions to talk about the probability that _at least one_ of the events happen.

- We use intersections to talk about the probability that _all_ events happen.

####Remark
In many cases, the probability of a set is jst the number of elements in the set dvided by the number of elements in the space. To emphasize we sometimes write $|A|$ for the number of elements of a set rather than $Pr(A)$.

Do a lot of simple problems rather than one hard problem.

###5.2.2 Statement of the Inclusion Exclusion Principle

####Definition (Inclusion-Exclusion)

Consider sets $A_1, A_2,..., A_n$. Denote the number of elements of a set $S$ by $|S|$ and the probability of a set $S$ by $Pr(S)$. Then (p. 181):

$$|\bigcup_{i=1}^n A_i| = \sum_{i=1}^n |A_i| - \sum_{i \le i \lt j \le n} |A_i \cap A_j| + \sum_{i \le i \lt j \lt k \le n} | A_i \cap A_j \cap A_k| ...$$ 


####Definition (Inclusion-Exclusion with Equally Likely Sets)
Assuming all sets $A_i$, $A_i \cap A_j ...$ have the same size:

$$|\bigcup_{i=1}^n A_i| = n|A_1| - \binom{n}{2}|A_{12}| + \binom{n}{3}|A_{123} - ... + (-1)^{n-1}|A_{12...n}|$$

###5.2.3 Justification of the Inclusion-Exclusion Formula

Proof based on _Binomial Theorem_, which says:

$$(x+y)^n = \sum_{k=0}^n \binom{n}{k}x^k y^{n-k}$$

See A.2.3 for a proof, and class notes (p. 183).

###5.2.4 Using Inclusion-Exclusion: Suited Hand

Look to class notes: idea is to make a large number of simple calculations rather than a single difficult calculation.

Consider a group of students, A, B, C, D who will be presenting. A does not want to go first, and D does not want to go last. Without restrictions, there are $4! = 24$ ways to arrange the students. The number of ways that A can be put into position 1 is $3! = 6$. The number of ways that D can be put into position 4 is $3! = 6$. The number of ways both A goes first and D goes last is $2! = 2$. So, we combine these values with inclusion-exclusion to get $24 - 6 - 6 + 2 = 14$.


###5.2.5 The At least to Exactly Method

####Definition (The At Least to Exactly Method)

Let $N(k)$ be the number of ways for at least $k$ things to happen and let $E(k)$ be the number of ways for exactly $k$ things to happen. Then:

$$E(k) = N(k) - N(k+1)$$

See p. 188 for application of this rule on counting suited hands.

##5.3 Derangements

A _derangement_ is an ordering such that no object is in its correct place.

###5.3.1 Counting Derangements

Let $A_i$ be the even that $i$ isn't moved. Then $A_1 \cup A_2 \cup ... \cup A_n$ is the even that at least one item occurs in the same place in the rearrangement. Note that all $A_i$ have the same size, since we can always relabel the elements without changing anything. Similarly all the intersections have the same size. Thus:

$$|\bigcup_{i=1}^n A_i| = n|A_i| - \binom{n}{2}|A_{ij}| + ... + (-1)^{n-1}|A_{12...n}|$$

- $|A_1| = (n-1)!$ as the first element is fixed, and so there are $n-1$ remaining elements to be permuted.

- $|A_{12}| = (n-2)!$ because of similar reasoning as above. Note that there are  $\binom{n}{2}$ such sets.

- $|A_{12...j}| = (n-j)!$, extending the same thought as above. There are $\binom{n}{j}$ such sets. Note that $\binom{n}{j} \cdot (n -j)! = \frac{n!}{j!}$.

So, inclusion-exclusion gives us:

$$|\bigcup_{i=1}^n A_i| = \frac{n!}{1!} - \frac{n!}{2!} + \frac{n!}{3!} + ... + (-1)^{n-1}\frac{n!}{n!}$$

This is the number of non-derangements.

Let $D_n$ denote the number of derangements. Then:

$$D_n = n! - \frac{n!}{1!} + \frac{n!}{2!} - \frac{n!}{3!} + ... + (-1)^{n-1}\frac{n!}{n!}$$

###5.3.2 The probability of a derangement

With some algebra, we see that the probability of a derangement is very close to $e^{-1}$; see p. 193.
